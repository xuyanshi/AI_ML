## Matplotlib练习

### 1. 中国的四个直辖市分别为北京市、上海市、天津市和重庆市，其2017年上半年的GDP分别为12406.8亿、13908.57亿、9386.87亿、9143.64亿。请绘制条形图来展示各自的GDP水平。

```python
import matplotlib.pyplot as plt
cities=['北京','上海','天津','重庆']
gdp=[12406.8,13908.57,9386.87,9143.64]
plt.barh(cities,gdp,color='blue')
plt.title('中国四个直辖市2017年上半年的GDP')
plt.xlabel('GDP（亿元）')
plt.ylabel('直辖市')
plt.tight_layout()
plt.rcParams['font.sans-serif']='SimHei'
plt.rcParams['font.family']='sans-serif'
plt.show()
```

### 2. 使用np.linspace()函数生成t∈[0, 2𝜋]。给定𝑥=16*sin⁡(𝑡)^3和𝑦=13*cos⁡(𝑡)−5*cos⁡(2𝑡)−2*cos⁡(3𝑡)−cos⁡(4𝑡)，使用红线画出x-y图表，并给图标添加一个题目：Heart

```python
import numpy as np
import matplotlib.pyplot as plt
t=np.linspace(0,2*np.pi,1000)
x=16*np.sin(t)**3
y=13*np.cos(t)-5*np.cos(2*t)-2*np.cos(3*t)-np.cos(4*t)
plt.figure(figsize=(8,6))
plt.plot(x,y,color="red",linewidth=2)
plt.title('Heart')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
