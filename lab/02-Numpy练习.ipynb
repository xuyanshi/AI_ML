### 3. ÂáΩÊï∞ f ÂÆö‰πâÂ¶Ç‰∏ãÔºåÂÖ∂‰∏≠xÊòØ‰ªªÊÑèÂÆûÊï∞: ùëì(ùë•)=(sin‚Å°(ùë•+1)/ùëí^ùë• )(ùë•+1)/2ÔºåÁîüÊàê‰∏Ä‰∏™ÂåÖÂê´5‰∏™ÂÖÉÁ¥†ÁöÑ‰∏ÄÁª¥Êï∞ÁªÑaÔºåËÆ°ÁÆób = f(a)„ÄÇ

```python
import numpy as np
def fx(n:np.ndarray) -> np.ndarray:
    """fx function."""
    return (np.sin(n+1)/np.exp(n))*(n+1)/2
a=np.random.rand(5)
b=fx(a)
print(a)
print(b)



```
